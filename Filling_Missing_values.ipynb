```python
df=df.replace('?',np.nan)
```

```python
df.isnull().sum().plot(kind='bar')
```

```python
a=df[df['normalized-losses']!='?']
b=(a['normalized-losses'].astype(int)).mean()
df['normalized-losses']=df['normalized-losses'].replace('?',b).astype(int)
```

```python
a=df[df['bore']!='?']
b=(a['bore'].astype(float)).mean()
df['bore']=df['bore'].replace('?',b).astype(float)
```

```python
a=df[df['stroke']!='?']
b=(a['stroke'].astype(float)).mean()
df['stroke']=df['stroke'].replace('?',b).astype(float)
```

```python
a=df[df['price']!='?']  #provides columns without '?'
b=(a['price'].astype(int)).mean() #converting a to int and finding mean 

df['price']=df['price'].replace('?',b)#repalcing '?',b
```

```python
a=df[df['horsepower']!='?']
b=(a['horsepower'].astype(int)).mean()
df['horsepower']=df['horsepower'].replace('?',b).astype(int)
```
